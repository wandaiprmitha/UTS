# One Hot Encode

```python
dt = data.drop(columns=['Surname'])
```

```python
dt = pd.get_dummies(dt)
```

```python
dt
```

# Normalisasi Data

```python
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
d = scaler.fit_transform(dt)
data_normalisasi = pd.DataFrame(d, columns=dt.columns)
data_normalisasi
```

# Membagi data menjadi X dan y

```python
X = data_normalisasi.drop(['Exited'],axis=1)
y = data_normalisasi['Exited']
```

```python
X.head()
```

```python
y.head()
```

# Train tanpa Split

```python
from sklearn.svm import SVC
model_ts = SVC()
```

```python
model_ts.fit(X,y)
```

```python
model_ts.score(X,y)
```

# Split Train dan Test Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

# Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
model_svm = SVC()
```

```python
model_svm.fit(X_train, y_train)
```

```python
model_svm.score(X_test,y_test)
```

# KFold SVM

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=5, shuffle=True, random_state=0)
```

```python
score_svm = cross_val_score(model_svm,X,y,cv=k_fold,n_jobs=1,scoring='accuracy')
print(score_svm)
```

```python
score_svm.mean()*100
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lr = LogisticRegression()
```

```python
model_lr.fit(X_train,y_train)
```

```python
model_lr.score(X_test,y_test)
```

# KFold Logistic Regression

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits = 10, shuffle = True, random_state = 0)
```

```python
score_lr = cross_val_score(model_lr,X,y,cv=k_fold,n_jobs=1,scoring='accuracy')
print(score_lr)
```

```python
score_lr.mean() * 100
```

# K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
model_knn.fit(X_train, y_train)
```

```python
model_knn.score(X_test, y_test)
```

# KFold KNN

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits = 100, shuffle = True, random_state = 0)
```

```python
score_knn = cross_val_score(model_knn,X,y,cv=k_fold,n_jobs=1,scoring='accuracy')
print(score_knn)
```

```python
score_knn.mean()*100
```
